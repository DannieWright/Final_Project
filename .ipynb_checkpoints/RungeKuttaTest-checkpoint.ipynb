```python
import RungeKutta as rk
import numpy as np
import pylab as plt
from scipy.integrate import odeint
%matplotlib inline
```

```python
def f_Newman_8_7_vec(r, t):
    (x,) = r
    r_prime = [-x**3+np.sin(t)]
    #print ("t {}, x {}, rp {}".format(t,x,r_prime))
    return np.array(r_prime)

if __name__ == "__main__":
    interval = (0.,10.)
    r0 = [0.]

    #set up plot
    #plot the step size vs. error
    plt.figure(figsize=(8, 5))
    ax = plt.subplot(111)

    #more efficient way to do this
    i=0

    ax.set_xlabel("$x$")
    ax.set_ylabel("$y$")
    ax.set_xlim(*interval)
    ax.set_ylim(-1.5,1.5)


    for order in [1, 2, 4]:
        # obtain solutions with different step sizes
        for steps in [20,1000]:  # Newman used [10,20,50,100]

            results = []    # to hold results with different step sizes

            solution = rk.vectorRungeKutta(f_Newman_8_7_vec, 1, r0, interval, steps, order=order)
            results.append(solution)

            # plot results

            if (order == 1):
                label = "Euler method"
            elif (order == 2):
                label = "Runge-Kutta method (2nd order)"
            elif (order == 4):
                label = "Runge-Kutta method (4th order)"

            if steps == 20:
                style='o'
            else:
                style='-'

            label = label + " steps=%d"%(steps)
            #for solution in results:
            plt.plot(solution[:,0],solution[:,1],style,label=label)

    plt.legend(bbox_to_anchor=(1.05, 1.05))
    plt.show()


            
```

```python
def lorenzRKScipy (steps, r0, state, interval):
    
    def lorenz (r, t):
        x, y, z = r

        fx = sigma * (y - x)
        fy = x * (rho - z) - y
        fz = x * y - beta * z

        return np.array ([fx, fy, fz], float)

    sigma, rho, beta = state

    order = 4
    numIndepVars = 3
    
    solutionRK = []
    solutionScipy = []
    diff = []
    
    
    solutionRK = rk.vectorRungeKutta (lorenz, numIndepVars, r0, interval, steps, order = 4)
    solutionScipy = odeint(lorenz, r0, solutionRK[:,0])
    diff = solutionRK[:,1:] - solutionScipy
    
    return solutionRK, solutionScipy, diff

interval = (0, 50)
steps = 100000
r0 = (1, 1, 1)
sigma = 10.0
rho = 28.0
beta = 8.0 / 3.0
state = (sigma, rho, beta)

solutionRk, solutionScipy, diff = lorenzRKScipy (steps, r0, state, interval)
```
